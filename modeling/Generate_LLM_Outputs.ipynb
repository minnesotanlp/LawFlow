## Helper Functions:

```python
import re

def extract_tags(response_text):
    """
    Extracts:
      - the conversation (or lawyer_action) block from the response, and
      - the next suggested step from the <next_suggested_step> tag.
    Returns a tuple: (conversation_text, next_step)
    """
    # Capture the entire block including the tags.
    convo_match = re.search(
        r"(<(?:conversation|lawyer_action)>.*?</(?:conversation|lawyer_action)>)",
        response_text, re.DOTALL | re.IGNORECASE
    )
    conversation_block = convo_match.group(1).strip() if convo_match else ""

    # Capture the next suggested step text.
    step_match = re.search(
        r"<next_suggested_step>(.*?)</next_suggested_step>",
        response_text, re.DOTALL | re.IGNORECASE
    )
    next_step = step_match.group(1).strip() if step_match else ""

    return conversation_block, next_step

# Rates below are for R1 model. For O1 model it is input_rate = 15 and output_rate=60 (Check website for updated costs)
def calculate_cost(completion, input_rate=0.035, output_rate=.550):
    """
    Calculates the cost based on the token usage from the API response per million tokens.

    Returns:
      input_cost, output_cost, iteration_cost
    """
    total_input_token = completion.usage.prompt_tokens
    total_output_token = completion.usage.completion_tokens

    input_cost = (total_input_token / 1_000_000) * input_rate
    output_cost = (total_output_token / 1_000_000) * output_rate
    iteration_cost = input_cost + output_cost
    return input_cost, output_cost, iteration_cost
```

```python
def prompt_and_update_context(gen_model, task, scenario, current_context, current_step):
    """
    Calls the model API with the generated prompt, extracts the required tags,
    and updates the context with the previous step and the full conversation/lawyer_action block.

    Returns:
      response_text, updated_context, next_step, completion
    """
    prompt = generate_prompt(task, scenario, current_context, current_step)

    completion = client.chat.completions.create(
    model = gen_model,
    messages=[
        {
            "role": "user",
            "content": [
                  {
                    "type": "text",
                    "text": prompt,
                  },
              ],
          }
      ],
    seed = 2025,
    temperature=0.7)


    response_text = completion.choices[0].message.content

    # Extract the conversation block (with tags) and next suggested step.
    extracted_convo, next_step = extract_tags(response_text)

    # Update the context with the previous step and the newly extracted conversation/lawyer_action block.
    updated_context = current_context + f"\n<step>\n{current_step}\n</step>\n\n<context>\n{extracted_convo}\n</context>\n"

    return response_text, updated_context, next_step, completion
```

### setup before running rest of the script

```python
current_step = "1.1.1 Take Notes" #Copy Paste this from above output

# First step
extracted_convo, _ = extract_tags(completion.choices[0].message.content)

next_step = "1.2 Identify potential complications/follow-up questions" #Copy Paste this from above output

# Update the context with the previous step and the newly extracted conversation/lawyer_action block.
context_str = f"\n<step>\n{current_step}\n</step>\n<context>\n{extracted_convo}\n</context>\n"

cumulative_cost = input_cost + output_cost
outputs_raw, llm_steps = [], []
outputs_raw.append(completion.choices[0].message.content)
llm_steps.append(current_step)
llm_steps.append(next_step)
```

Run steps 2 to the end:

```python
for i in range(50):
      print(f"\n--- Iteration {i+1} ---")

      response_text, context_str, next_step, completion = prompt_and_update_context(
          gen_model, task_diagram_44, SCENARIO, context_str, next_step)

      # Store LLM outputs
      outputs_raw.append(response_text)
      llm_steps.append(next_step)

      # Calculate cost
      input_cost, output_cost, iteration_cost = calculate_cost(completion)
      cumulative_cost += iteration_cost

      # print(response_text)
      # print("")

      # Terminate loop if next suggested step is "END"
      if next_step.strip().upper() == "END":
          print("Next suggested step is 'END'. Terminating loop.")
          break

print(f"\nTotal cumulative cost: {cumulative_cost:.6f}")
```
